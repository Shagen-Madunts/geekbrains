> Даны три прямые $$y=k_1x+b_1$$ $$y=k_2x+b_2$$ $$y=k_3x+b_3$$ Как узнать, пересекаются они в одной точке или нет?

> **Ответ**: 
>Прямые пересекаются, если:
- угловые коээфициенты не равны друг другу
$$k_1 \neq k_2 \neq k_3 $$ 
- и равны констаны b
$$b_1=b_2=b_3$$

# Задание 3

> На листе тетради «в линейку» (расстояние между линиями равно $а$) лежит игла (длиной $b$). Координаты нижней точки иглы (х,у), 
игла лежит под углом $\alpha$. Пересекает ли игла линию или нет?

> **Ответ**: 
Допустим, что координаты нижней точки иглы $x$ и $y$ лежат на одной из линий тетради. Получаем прямоугольный треугольник, где игла - это гипотенуза, $\alpha$ - угол напротив $a$. Т.о. если $\frac{a}{\sin\alpha} \leq b$ и $\alpha \neq 0$, то игла  пересекает линию. В общем случае если начало иглы не лежит на одной из прямых тетради, то должно выполняться условие: $\frac{(a-y)}{\sin\alpha} \leq b$
