# Задание 3

- Напишите код на Python, реализующий построение графиков: окружности, эллипса, гиперболы.

```python
%matplotlib notebook

"""График окружности"""
x_list_pos = []
x_list_neg = []
y1 = []
y2 = []
y1_ = []
y2_ = []
r2 = 100 # квадрат радиуса
list_num = [] # будущий список для x значений
k = 0.01
while k < m.sqrt(r2):
    list_num.append(k)
    k += 0.01

for x in list_num:
    xpos = x
    xneg = -x
    x_list_pos.append(xpos)
    x_list_neg.append(xneg)
    y1.append(m.sqrt(r2 - xpos**2)) 
    y2.append(m.sqrt(r2 - xneg**2)) 
    y1_.append(-m.sqrt(r2 - xpos**2)) 
    y2_.append(-m.sqrt(r2 - xneg**2))  
plt.plot(x_list_pos,y1)
plt.plot(x_list_neg,y2)
plt.plot(x_list_pos,y1_)
plt.plot(x_list_neg,y2_)
plt.xlabel("x")
plt.ylabel("y")
plt.axes().set_aspect(1)
```

```python
%matplotlib notebook
"""График элипса"""
x_list_pos = []
x_list_neg = []
y1 = []
y2 = []
y1_ = []
y2_ = []
a = 2
b = 10 
list_num = [] # будущий список для x значений
k = 0.01
while k < m.pi:
    list_num.append(k)
    k += 0.01

for x in list_num:
    xpos = a * np.cos(x)
    xneg = -a * np.cos(x)
    x_list_pos.append(xpos)
    x_list_neg.append(xneg)
    y1.append(b * np.sin(x)) 
    y2.append(b * np.sin(x)) 
    y1_.append(-b * np.sin(x)) 
    y2_.append(-b * np.sin(x))  
plt.plot(x_list_pos,y1)
plt.plot(x_list_neg,y2)
plt.plot(x_list_pos,y1_)
plt.plot(x_list_neg,y2_)
plt.xlabel("x")
plt.ylabel("y")
```

```python
%matplotlib notebook
"""График гиперболы"""
x_list_pos = []
x_list_neg = []
y1 = []
y2 = []
y1_ = []
y2_ = []
a = 1
b = 10 
list_num = [] # будущий список для x значений
k = 1.0001
while k < m.sqrt(5):
    list_num.append(k)
    k += 0.01

for x in list_num:
    xpos = x
    xneg = -x
    x_list_pos.append(xpos)
    x_list_neg.append(xneg)
    y1.append(m.sqrt(b**2 * xpos**2 / a**2 - b**2)) 
    y2.append(m.sqrt(b**2 * xneg**2 / a**2 - b**2)) 
    y1_.append(-m.sqrt(b**2 * xpos**2 / a**2 - b**2))
    y2_.append(-m.sqrt(b**2 * xpos**2 / a**2 - b**2))  
plt.plot(x_list_pos,y1)
plt.plot(x_list_neg,y2)
plt.plot(x_list_pos,y1_)
plt.plot(x_list_neg,y2_)
plt.xlabel("x")
plt.ylabel("y")
```

- **Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.**

По сути требуется доказать, что при переводе точек $$ T_1(x_1,y_1)$$ 
$$T_2(x_2,y_2)$$ 
через ортогорнальное преобразование в точки 
$$ T_1'(X_1,Y_1)$$
$$T_2'(X_2,Y_2)$$ 
их отрезки имеют равные длины. Т.е.  $$ |T_1T_2| = |T_1'T_2'|$$

$$ |T_1'T_2'|^2 = (X_2 - X_1)^2 + (Y_2-Y_1)^2 = (a_{11}(x_2-x_1) + a_{12}(y_2-y_1))^2 + (a_{21}(x_2-x_1) + a_{22}(y_2-y_1))^2 = (x_2-x_1)^2 + (y_2-y_1)^2 = |T_1T2_2|^2 $$

```python
"""
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1

"""
from scipy.optimize import fsolve
%matplotlib notebook

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
f1 = lambda x: x**2 - 1 
f2 = lambda x: (np.exp(x) + x - 1) / x

plt.plot(x,f1(x))
plt.plot(x,f2(x))
plt.ylim(-10,30)
plt.xlim(-3,5)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

def decision(coord):
    x, y = coord
    return (x**2-1-y, np.exp(x) + x*(1-y)-1)

x1, y1 =  fsolve(decision, (-1, 0))   
print(f"Одно из решений системы уравнений: {x1}, {y1}")
```
